```python
import jieba
```

### Method 1

```python
INPUT_FILE = '/content/drive/MyDrive/Corpus/chinese/thesis data chinese.txt'
OUTPUT_FILE = '/content/drive/MyDrive/Corpus/chinese/thesis data chinese segmented.txt'

# Load all lines
with open(INPUT_FILE, "r", encoding="utf-8") as f:
    lines = f.readlines()

# Total must be 24,000 as you said
print("Total lines:", len(lines))

def segment_and_save(lines, output_file):
    with open(output_file, "w", encoding="utf-8") as f:
        for line in lines:
            line = line.strip()
            if not line:
                f.write("\n")
                continue

            segmented = " ".join(jieba.cut(line))
            f.write(segmented + "\n")

# Segment and write
segment_and_save(lines, OUTPUT_FILE)

print("Done! Files created")
```

### Method 2

```python
INPUT_FILE = '/content/drive/MyDrive/Corpus/ucsy_news_ch.txt'

TRAIN_FILE = '/content/drive/MyDrive/Corpus/chinese/train-ch.txt'
VAL_FILE = '/content/drive/MyDrive/Corpus/chinese/valid-ch.txt'
TEST_FILE = '/content/drive/MyDrive/Corpus/chinese/test-ch.txt'

# Load all lines
with open(INPUT_FILE, "r", encoding="utf-8") as f:
    lines = f.readlines()

# Total must be 24,000 as you said
print("Total lines:", len(lines))

# Split according to your rules
test_lines = lines[:3000]
val_lines = lines[-3000:]
train_lines = lines[3000:-3000]

def segment_and_save(lines, output_file):
    with open(output_file, "w", encoding="utf-8") as f:
        for line in lines:
            line = line.strip()
            if not line:
                f.write("\n")
                continue

            segmented = " ".join(jieba.cut(line))
            f.write(segmented + "\n")

# Segment and write
segment_and_save(train_lines, TRAIN_FILE)
segment_and_save(val_lines, VAL_FILE)
segment_and_save(test_lines, TEST_FILE)

print("Done! Files created:")
print(" -", TRAIN_FILE)
print(" -", VAL_FILE)
print(" -", TEST_FILE)
```
