```python
import pandas as pd
import os
```

```python
ch_path = '/content/drive/MyDrive/Corpus/chinese/thesis data chinese segmented.txt'
my_path = '/content/drive/MyDrive/Corpus/chinese/thesis data segmented.txt'
```

```python
try:
    ch_file = open(ch_path, 'r', encoding='UTF-8')
    my_file = open(my_path, 'r', encoding='UTF-8')
except FileNotFoundError as fe:
    print(f"File was not found: {fe}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
ch = []
my = []

for line in ch_file:
    ch.append(line.strip())

for line in my_file:
    my.append(line.strip())
```

```python
data = {'ch': ch, 'my': my}
df = pd.DataFrame(data=data)
```

```python
df.to_excel('/content/drive/MyDrive/Corpus/chinese/thesis_data.xlsx', index=False)
```
