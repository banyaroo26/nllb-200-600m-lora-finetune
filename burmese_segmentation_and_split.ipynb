```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
```

```python
# change the folder name from `myanmar-tokenizer` to `myanmar_tokenizer`
from myanmar_tokenizer import p3_tokenizer
```

```python
tokenizer = p3_tokenizer.MyanmarTokenizer(separator=" ")
```

```python
text = 'မနက်ခင်းလေနုအေးကလန်းဆန်းနေ၏။'
```

```python
result = tokenizer.cut(text)
result
```

```python
en_path = '/content/drive/MyDrive/Corpus/open/thesis data en.txt'
my_path = '/content/drive/MyDrive/Corpus/open/thesis data segmented.txt'
## out_path = '/content/drive/MyDrive/Corpus/segmented.txt'
```

```python
try:
  en_file = open(en_path, 'r', encoding='UTF-8')
  my_file = open(my_path, 'r', encoding='UTF-8')
  ## out_file = open(out_path, 'w', encoding='UTF-8')
except FileNotFoundError as fe:
    print(f"File was not found: {fe}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
en = []
my = []

for line in en_file:
  en.append(line.strip())

for line in my_file:
  my.append(line.strip())
```

```python
data = {'en': en, 'my': my}
df = pd.DataFrame(data=data)
```

```python
df.sample(5)
```

```python
df['my'] = df['my_original'].apply(tokenizer.cutRecursively)
```

```python
df.sample(5)
```

```python
df.to_excel('/content/drive/MyDrive/Corpus/open/thesis_data.xlsx')
```

```python
new_df = df.copy()
new_df.drop(columns=['my_original'], inplace=True)

train_df, val_test = train_test_split(new_df, train_size=0.8, random_state=11)
val_df, test_df    = train_test_split(val_test, train_size=0.5, random_state=11)
```
